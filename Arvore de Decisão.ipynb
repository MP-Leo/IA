```python
import pandas as pd
```

```python
df = pd.read_csv('heart.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df['output']
df.drop(columns=['output'], inplace=True)
```

```python
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.3, random_state=0, stratify=y)
```

```python
y = y_train
y_str = y.astype('category')
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=6)
clf.fit(X_train, y_train)
```

```python
preds = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, preds)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(24,11))
plt.title('teste')
tree.plot_tree(clf, label=None, node_ids=True, rounded=True, fontsize=9, feature_names=X_train.columns, filled=True, class_names=["Teve ataque cardiaco", "Nao teve"], proportion=True)
# plt.savefig('decision_tree.png', format='png')
plt.show()
```
